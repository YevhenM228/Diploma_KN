```python
import os
from google.colab import drive
import time
# 1. Підключаємо Google Drive (якщо ще не підключено)
if not os.path.exists('/content/drive'):
    drive.mount('/content/drive')
else:
    print("Google Drive вже підключено.")

# 2. Визначаємо базову папку для бекапу на Google Drive
# Рекомендую дати їй унікальне ім'я, наприклад, з поточною датою/часом
timestamp = time.strftime("%Y%m%d-%H%M%S") # Потрібно імпортувати time зверху
project_backup_dir = f'/content/drive/MyDrive/YOLOv11_Colab_Backup_{timestamp}/'
os.makedirs(project_backup_dir, exist_ok=True)
print(f"Створено папку бекапу: {project_backup_dir}")

# 3. Копіюємо основні папки та файли

# Копіюємо папку my_model (якщо вона є і містить важливі речі)
if os.path.exists('/content/my_model'):
    !cp -r /content/my_model/ "{project_backup_dir}my_model/"
    print("Папку 'my_model' скопійовано.")

# Копіюємо папку runs/ (з результатами тренування)
if os.path.exists('/content/runs'):
    !cp -r /content/runs/ "{project_backup_dir}runs/"
    print("Папку 'runs' (з результатами тренування) скопійовано.")
else:
    print("Папка 'runs' не знайдена в /content/. Можливо, результати тренування вже на Drive.")

# Копіюємо окремі файли, які видно на скріншоті
files_to_copy = [
    'dataset.yaml',
    'my_model.zip',
    'yolov11l.pt' # Якщо ти хочеш зберегти цю конкретну модель
]

for filename in files_to_copy:
    source_path = os.path.join('/content/', filename)
    dest_path = os.path.join(project_backup_dir, filename)
    if os.path.exists(source_path):
        !cp "$source_path" "$dest_path"
        print(f"Файл '{filename}' скопійовано.")
    else:
        print(f"Файл '{filename}' не знайдено в /content/.")

print("\nУсі вказані файли та папки скопійовано на Google Drive!")
print(f"Зверніться до папки: {project_backup_dir}")
```
